```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
```

```python
df
```

```python
# Cleaning Last Name column 

df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
        
```

```python
# Drop duplicates

df = df.drop_duplicates()
df
```

```python
# Drop columns because is not useful. 

df.drop(columns = 'Not_Useful_Column')
```

```python
# Cleaning format for Phone Number colum . 

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]' , '')
df
```

```python
# Trasform Phone Number column in relevant format for phone numbers . 

df['Phone_Number'] = df['Phone_Number'].apply( lambda x : str(x))
df['Phone_Number'] = df['Phone_Number'].apply( lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
# Replacing nan & na values with blank space
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', " ")
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', " ")
df
```

```python
# Spliting the Adress column to 3 different columns . 

df[['Street_Adress','State','Zip_Code']] = df['Address'].str.split(',',2 , expand = True)
df
```

```python
# drop address column after spliting

df.drop(columns = 'Address')
```

```python
# Replacing Yes with Y and No with N and Nan & Na with blank values

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')


df
```

```python
# filling Nan and N.a with blank values 

df = df.fillna('')
df
```

```python
# Filtering out just contact that can be called 

for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x ,inplace = True)
df
```

```python
# Drop rows with missing phone number because they can be called . 

for x in df.index:
    if df.loc[x,'Phone_Number'] == ' ':
        df.drop(x ,inplace = True)
df
```

```python
# reset index with normal one

df = df.reset_index(drop = True)
df
```
